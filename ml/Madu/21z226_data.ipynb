**Encoding categorical values into numerical values**

```python
#date: We utilize a constant reference date called 'current_date.'
#Subsequently, we compute the elapsed number of days from the 'current_date' to each date within the 'Date' column.
#These calculated values replace the existing column values.
ml['Date'] = pd.to_datetime(ml['Date'])
print(ml['Date'].head())

current_date = pd.Timestamp('2024-02-21')
ml['Date'] = (current_date - ml['Date']).dt.days

print(ml['Date'].head())
```

```python
#The dataset contains 54 unique regions. To avoid introducing unintended correlations between variables, one-hot encoding is employed instead of label encoding.
# The OneHotEncoder is utilized to generate a one-hot encoded matrix named 'region_encoded'.
#This matrix is organized into a DataFrame containing newly created column names.
# Following this transformation, the encoded matrix is added to the original dataset. Consequently, the original 'region' column is removed.
#At this stage, the dataset expands to encompass 67 columns.
region_counts = ml['region'].nunique()
print(region_counts)
```

```python
encoder = OneHotEncoder(sparse_output=False)

region_encoded = encoder.fit_transform(ml[['region']])
region_encoded_df = pd.DataFrame(region_encoded, columns=encoder.get_feature_names_out(['region']))

ml = pd.concat([ml, region_encoded_df], axis=1)
ml.drop('region', axis=1, inplace=True)

print(ml.head())
```

**counting the number of outliers present in each column**

```python
def count_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers.sum()

outliers_counts = {}
for column in ml.columns:
    outliers_counts[column] = count_outliers(ml[column])

for column, count in outliers_counts.items():
    print(f"Number of outliers in {column}: {count}")
```

**HANDLING OUTLIERS**

```python
def replace_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    column = column.apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
    return column

columns_to_replace_outliers = ['Total Volume', '4046', '4225', '4770', 'Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags']

for column in columns_to_replace_outliers:
    ml[column] = replace_outliers(ml[column])

print(ml)
```

Removing:

=> AveragePrice has less no. of outliers. As removing them won't cause any information loss, outliers in this column are removed. **bold text**

```python
def remove_outliers(dataset, column_name):
    Q1 = column_name.quantile(0.25)
    Q3 = column_name.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_rows = (column_name < lower_bound) | (column_name > upper_bound)

    dataset = dataset[~outlier_rows]
    return dataset

columns_to_remove_outliers = ['AveragePrice']

for column in columns_to_remove_outliers:
    ml = remove_outliers(ml, ml[column])

ml.reset_index(drop=True, inplace=True)

print(ml)
```

Feature Scaling

Normalization:

=> It is one of the ways to do Feature Scaling.

=> Normalization is used here to bring all the variables into the same range or same scale (0-1) to avoid dominance of one variable over the other.

=> All the columns except the dependent variable is normalized.

```python
from sklearn.preprocessing import Normalizer

norm = Normalizer()
dependent_variable = 'AveragePrice'
independent_variables = ml.columns.drop(dependent_variable)
n_data = norm.fit_transform(ml[independent_variables])
n_dataset = pd.DataFrame(n_data, columns=independent_variables)
n_dataset = pd.concat([n_dataset, ml[dependent_variable]], axis=1)

print(n_dataset.head())
print(n_dataset.describe())
n_dataset.to_csv('avocado2.csv', index=False)
```

Splitting of Training data and Testing data

The normalized dataset is splitted into X_train and X_test.80% is used for training and 20% is used for testing.

```python
from sklearn.model_selection import train_test_split

X = n_dataset
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

The training and testing data are being stored in separate .csv files.

```python
X_train.to_csv('train_data.csv', index=False)
X_test.to_csv('test_data.csv', index=False)
```
